```python
# identity_metric * constant_k
# n : len of metric

def multi_i(k, n):
    C = []
    for i in range(0, n):
        tmp = []
        for j in range(0, n):
            if i==j:
                tmp.append(k*1)
            else:
                tmp.append(0)
        C.append(tmp)
    return C
```

```python
# metric_A * constant_k
# l : len of A

def multi_k(k, A, l):
    C = []
    for i in range(0, l):
        tmp = []
        for j in range(0, l):
            tmp.append(k*A[i][j])
        C.append(tmp)
    return C
```

```python
# metric_A - metric_B
# l : len of A, len of B

def subtr(A, B, l):
    C = []
    for i in range(0, l):
        tmp = []
        for j in range(0, l):
            tmp.append(A[i][j]-B[j][i])
        C.append(tmp)
    return C
```

```python
# metric_A * metric_B

def multi(A, B):
    l = len(A)
    C = []
    for i in range(0, l):
        tmp = []
        for j in range(0, l):
            tmp.append(A[i][j]*B[j][i])
        C.append(tmp)
    return C
```

```python
def Trace(A):
    sum = 0
    l = len(A)
    for i in range(l):
        sum = sum + A[i][i]
    return sum
```

```python
def RamanujanGraph(A, k):
    
    n = len(A)
    
    q = -1
    for i in range(n):
        q = q + A[0][i]
    
    
    # L = INDICES(k)
    L = []
    L.append(k)
    while (k%2==0):
        k = k/2
        L.append(k) 
    while (k!=1):
        k = (k+1)/2
        L.append(k)
        L.append(k-1)
        if (k%2==0):
            k = k-1
    
    l = len(L)
    
    # let T[0..3] denote a new array with T[0] = A
    T = []
    T.append(A)
    B = [[0 for i in range(n)] for i in range(n)]
    T.append(B)
    T.append(B)
    T.append(B)
    
    # let Q[0..3] denote a new array with Q[0] = 1
    Q = []
    Q.append(1)
    Q.append(0)
    Q.append(0)
    Q.append(0)
    
    
    start_time = time.time()
    
    for i in range(l-1, 0, -1):
        for j in range(3, 0, -1):
            Q[j] = Q[j-1]
            T[j] = T[j-1]
            
        if (L[i-1]%2 == 0):
            if (L[i-1] == 2*L[i]):
                j = 1
            else:
                j = 2
            
            if (L[i+j-1]%2 == 0):
                Q[0] = Q[j] * Q[j]
            else:
                Q[0] = q * Q[j] * Q[j]
            
            #T[0] = T[j]*T[j]-2Q[0]In    #18
            TMP1 = multi(T[j], T[j])
            TMP2 = multi_i(2*Q[0], n)
            T[0] = subtr(TMP1 , TMP2, n)
            
            #trace(T[0]) > (Q[0]−1)^2 + 2nQ[0]    #19
            if Trace(T[0]) > ( (Q[0]-1)*(Q[0]-1) + 2*n*Q[0] ):
                end_time = time.time()
                print("Execution time : %s seconds" % (end_time - start_time))
                return False
            
        else:
            if ( L[i-1] == (2*L[i+1]-1) ):
                j = 2
            else:
                j = 1
                
            Q[0] = Q[j]*Q[j+1]
             
            #T[0] = T[j]*T[j+1]-Q[0]A    #24
            TMP3 = multi(T[j], T[j+1])
            TMP4 = multi_k(Q[0], A, n)
            T[0] = subtr( TMP3, TMP4, n)

    end_time = time.time()
    print("Execution time : %s seconds" % (end_time - start_time))
    return True
```
